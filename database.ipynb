### Initializing the database and populating it

In this step, I initialize the `Agent` and populate the pattern database with Q-values for various state-action pairs. The Q-values are assigned based on the distance of the state from the initial state, following a predefined pattern. The populated Q-values are then saved to a file named `pattern_database.pkl` using the `pickle` module. 



```python
agent = Agent()
agent.register_patterns()

with open('pattern_database.pkl', 'wb') as f:
    pickle.dump(agent.QV, f)

print("Pattern database initialized, populated, and saved to 'pattern_database.pkl'.")
```
